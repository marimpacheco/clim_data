```python
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_context('talk')
```

```python
die = np.arange(1, 7, 1)
```

```python
population_mean = die.mean()
print("Population mean: {}".format(population_mean))
```

```python
sns.distplot(die, kde=False, norm_hist=True, bins=np.arange(0.5, 7, 1), label="PDF");
plt.xlabel("Pips");
plt.ylabel("Probability");
plt.vlines(die.mean(), 0, 0.17, label="mean");
plt.legend(loc=2);
plt.title("Throwing a fair dice")

#plt.savefig('../img/2_parameter_statistic_1.svg', bbox_inches='tight', pad_inches=.1)
```

```python
#np.random.seed(0)

N = 5
sample = np.random.choice(die, N)
sns.distplot(sample, kde=False, norm_hist=True, bins=np.arange(0.5, 7, 1), label="ED");
plt.xlabel("Pips");
plt.ylabel("Probability");
plt.vlines(sample.mean(), 0, 0.17, label="sample mean");
plt.legend(loc=2);
plt.title("N = {}".format(N));

#plt.savefig('../img/2_parameter_statistic_2.svg',  bbox_inches='tight', pad_inches=.1)
```

```python
#np.random.seed(0)

rep = 10000 #good enough value without waiting too long
N = 10

sample_mean = []
for i in range(rep):
    sample=np.random.choice(die, N)
    sample_mean.append(sample.mean())

sns.distplot(sample_mean, kde=False, norm_hist=True);
plt.title("N = {}".format(N));

# discrete population, our number of possible outcomes is also discrete
```

```python
np.random.seed(0)

rep= 10000
N = 100 # if increase N better approx the stat mean with the population mean

sample_mean = []
for i in range(rep):
    sample_mean.append(np.random.choice(die, N).mean())

sns.distplot(sample_mean, kde=False, norm_hist=True);
#plt.title("N = {}".format(N));
```
