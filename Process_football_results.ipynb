```python
## Process the raw data from Swedish Men's football league, Allsvenskan 2019
## Put into a CSV file for easy reading

import numpy as np
def read_tabell(file):
    # Open file
    fin = open(file,'r')
    # Read all lines
    lines = fin.readlines()
    
    # Output file
    fut = open('allsvenskan_2019.csv','w')
    # Output header
    fut.write('%s,%s,%s,%s\n' % ('HomeTeam','AwayTeam','HomeTeamGoals','AwayTeamGoals'))
    # List of all written lines
    lines_out = []
    
    # First line has all away teams
    away_teams = lines[0].split()
    
    # Start on second line
    for line in lines[1:]:
        # First element is home team, then comes results
        tmp = line.split()
        home_team = tmp[0]
        
        ii = 0
        # Results 
        for res in tmp[1:]:
            # For invalid results, i.e. AIK vs AIK, the result is -
            # otherwise it is 2-1 or similar
            # If we can split string into two elements, it is a result
            # otherwise it is invalid
            try:
                # Home goals and away goals
                hg,ag = res.split('–')
                lineout = '%s,%s,%s,%s' % (home_team,away_teams[ii],hg,ag)
                print(lineout)
                lines_out.append(lineout)
                # Write to file
                fut.write('%s,%s,%s,%s\n' % (home_team,away_teams[ii],hg,ag))
            
            except:
                # Control print invalid results to check
                print(' not valid result between %s and %s: %s ' % (home_team,away_teams[ii],res) )
            
            ii = ii+1
            
    return lines_out

# Call read function
lines = read_tabell('allsvenskan_2019_dont_use.txt')
```
