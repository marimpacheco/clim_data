```python
import numpy as np
import matplotlib.pyplot as plt
data_file = 'nino34.long.anom.data.txt'
data = np.genfromtxt( data_file, skip_header=1, skip_footer=7 )
years = data[:, 0]
nino34 = data[:, 1:].flatten()
time = np.arange(years[0] + 1 / 24, years[-1] + 1, 1 / 12 )
mask = (nino34 > -50.) & (time >= 1901) & (time < 2014)
nino34 = nino34[mask]
time = time[mask]

data_file = "all_india_rainfall.csv"
data = np.genfromtxt(data_file, skip_header=1, delimiter='","', deletechars='"')
rain = data[:, 1:13].flatten()
rain_clim = data[:, 1:13].mean(axis=0)
rain_anom = (data[:, 1:13] - rain_clim).flatten()

fig,ax=plt.subplots(1,1)
ax.plot(nino34)
ax.plot(rain_anom)
```
